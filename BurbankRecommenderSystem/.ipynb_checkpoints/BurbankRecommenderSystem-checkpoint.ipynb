```python
def BlockType(s):
    s = s.strip().lower()
    if s == "irregular":
        return 1
    if s == "regular":
        return 2
    if s == "corner":
        return 3
    return 0
def Storey(s):
    s = s.strip().lower()
    if s == "single":
        return 1
    if s == "double":
        return 2
    return 0
```

```python
import pandas as pd
import scipy.stats as ss
from sklearn.preprocessing import StandardScaler
```

```python
#ata = pd.read_csv('BDataFile.csv',encoding='utf8')
data = pd.read_csv('BDataFile.csv',encoding='utf8',converters={"BlockType":BlockType,"Storey":Storey})

#storey = set(data['Storey'])
#data
data2=data.insert(0,[0,'test','test','test',1,2,673511,0,1,3,2,508,3,148])
print(data2)

df = pd.DataFrame(data, columns=['Storey', 'CarSpaces', 'Price', 'Study', 'Ensuite','BedRooms','LandSize','BlockType','HouseSize']).astype('float64')

sc_X = StandardScaler()
df = sc_X.fit_transform(df)
df

```

```python
num_cols = len(df[0,:])
for i in range(num_cols):
    col = df[:,i]
    col_stats = ss.describe(col)
    print(col_stats)
```
