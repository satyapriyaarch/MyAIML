```python
def BlockType(s):
    s = s.strip().lower()
    if s == "irregular":
        return 1
    if s == "regular":
        return 2
    if s == "corner":
        return 3
    return 0
def Storey(s):
    s = s.strip().lower()
    if s == "single":
        return 1
    if s == "double":
        return 2
    return 0
```

```python
import pandas as pd
import numpy as np
import scipy.stats as ss
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
data1 = pd.read_csv('BDataFile.csv',encoding='utf8',converters={"BlockType":BlockType,"Storey":Storey})
test = pd.read_csv('BTest.csv',encoding='utf8',converters={"BlockType":BlockType,"Storey":Storey})

data3 = pd.merge(test,data1,how='outer')
data3.iloc[0:,4:]
```

```python
num_cols = len(df[0,:])
for i in range(num_cols):
    col = df[:,i]
    col_stats = ss.describe(col)
    print(col_stats)
```

```python
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(data1.iloc[0:,4:], data1['Packageid'])
print(neigh.predict(test.iloc[0:,4:]))
print(neigh.kneighbors(test.iloc[0:,4:])[1])
```

```python
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(dfx[1:,4:], data1['Packageid'])
print(neigh.predict(dfx[0:1,4:]))
print(neigh.kneighbors(dfx[0:1,4:])[1])

results = neigh.kneighbors(dfx[0:1,4:])[1]
print(results)
```
